###TASK 3B: Show all registered members:

```python



# Display all books with more than 3 available copies

mem ="SELECT* FROM members"
df = pd.read_sql(mem, conn)
df.head()
```

Find all loans made on a specific date

```python
loan_date ="""
SELECT*
FROM loans
WHERE loan_date >= '2024-11-15'
"""
df = pd.read_sql(loan_date, conn)
df.head()
```

Display all books with more than 3 available copies

```python
bk_copy ="""
SELECT*
FROM books
WHERE available_copies >= 3
"""
df = pd.read_sql(bk_copy, conn)
df.head()
```

# **Phase 4: Applying Comparison and Logical Operators (Day 4)**


Books published after 2015:


```python
bk_published ="""
SELECT*
FROM books
WHERE published_year > 2015
"""
df = pd.read_sql(bk_published, conn)
df.head()
```


Members who joined before 2022



```python
mem_joined ="""
SELECT*
FROM members
WHERE join_date < '2022-01-01'
"""
df = pd.read_sql(mem_joined, conn)
df.head()
```

Books that are either Fiction or Mystery

```python
bk_genre="""
SELECT*
FROM books
WHERE genre = 'Fiction' OR genre = 'Mystery'
"""
df = pd.read_sql(bk_genre, conn)
df.head()
```

Loans that have not been returned

```python
is_returned="""
SELECT*
FROM  loans
WHERE returned =True
"""
df = pd.read_sql(is_returned, conn)
df.head()
```

**Members who joined after 2020 and have borrowed books (*bonus*)**

```python
# ADDS member_id to the member table
%%sql
ALTER TABLE members
ADD COLUMN member_id INTEGER;

-- Step 2: Update existing records with sequential numbers
UPDATE members
SET member_id = (SELECT COUNT(*) FROM members m2 WHERE m2.rowid <= members.rowid);
```

```python
# Members who joined after 2020 and have borrowed books (bonus)
select_mem ="""
  SELECT
    m.member_id,
    m.first_name,
    m.last_name,
    l.loan_date,
    l.return_date,
    l.returned
  FROM  loans as l
  INNER JOIN members as m
  ON l.member_id = m.member_id
  WHERE m.join_date > '2020-12-31'
"""
df = pd.read_sql(select_mem, conn)
df.head()
```

# **Phase 5: Sorting, Filtering, and Limiting Results (Day 5 - ORDER BY, DISTINCT, OFFSET, LIMIT)**


Sort all books alphabetically by title

```python
sort_asc = """
SELECT*
FROM  books
ORDER BY title ASC
"""
df = pd.read_sql(sort_asc, conn)
df.head()
```

List distinct book genres

```python
get_distinct = """
SELECT DISTINCT(genre),
  b.author,
  b.title
FROM books  as b
"""
df = pd.read_sql(get_distinct, conn)
df.head()

```

Get top 5 most recent members

```python
top_recent_mem = """
SELECT*
FROM  members
ORDER BY join_date DESC
LIMIT 5
"""
df = pd.read_sql(top_recent_mem, conn)
df.head(10)
```

Skip the first 5 books and display the next 5


```python
skip_bk =  """
 SELECT *
 FROM books
 LIMIT 5
 OFFSET 5

"""
DF = pd.read_sql(skip_bk, conn)
DF.head()
```

Show the 10 most recent loans not yet returned

```python
recent_not_returned = """
SELECT*
FROM  loans
WHERE returned =False
ORDER BY loan_date DESC
LIMIT 10
"""
df = pd.read_sql(recent_not_returned, conn)
df.head(10)
```
